```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from statistics import mean
```

```python
res = requests.get("https://gist.github.com/paulmillr/2657075")
```

```python
html_doc = res.text
```

```python
soup = BeautifulSoup(html_doc, "html.parser")
```

```python
table = soup.find("table")
```

```python
contributors = table.findChildren("tr")
```

```python
contributors_name = [contributor.find("a").text for contributor in contributors[1:]]
```

```python
r = requests.get("https://api.github.com/users/"+contributors_name[4]+"/repos")
```

```python
import json
```

```python
d={}
```

```python
for usr in contributors_name[177:]:
    r = requests.get("https://api.github.com/users/"+usr+"/repos")#, auth=(username,token))

    json_res = json.loads(r.text)

    len(json_res)

    stars = [j["stargazers_count"] for j in json_res]
    try:
        d[usr] = mean(stars)
    except:
        d[usr] = 0
```

```python
data = pd.DataFrame(data = {"usr": list(d.keys()), "stars":list(d.values())})
```

```python
data["stars"] = data.stars.apply(lambda x : float("{:.1f}".format(x)))
```
